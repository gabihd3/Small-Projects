```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
import pickle
```

```python
import os 
path = ('/Users/gabriel/Documents/data')
os.chdir(path)
```

```python
df = pd.read_csv('Last_preprocessing.csv')
df.head()
```

```python
inputs = df.iloc[:,:-1]
inputs.head()
```

```python
targets = df['Targets']
targets.head()
```

```python
x_train,x_test,y_train,y_test = train_test_split(inputs,targets,train_size = 0.8, random_state = 20)
```

```python
print(x_train.shape,y_train.shape)
print(x_test.shape,y_test.shape)
```

```python
reg = LogisticRegression()
```

```python
reg.fit(x_train,y_train)
```

```python
print('Score prediction:',reg.score(x_train,y_train).round(3)*100)
```

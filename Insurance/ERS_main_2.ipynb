# Getting Dummies

```python
a0 = categorical_df.drop(['V8','PM1','C1','C2'],axis = 1)
dum_a0 = pd.get_dummies(a0,prefix = None)
dum_a0.head()
```

```python
v8 = categorical_df.drop(['A0','PM1','C1','C2'],axis = 1)
dum_v8 = pd.get_dummies(v8)
dum_v8.head()
```

```python
c2 = categorical_df.drop(['A0','PM1','C1','V8'],axis = 1)
dum_c2 = pd.get_dummies(c2)
dum_c2.head()
```

# Scaling Numerical Data

```python
numerical_df = df.iloc[:,5:]
numerical_df.head(3)
```

```python
numerical_df.columns
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
scaler.fit(numerical_df)
```

```python
scaled_numerical = scaler.transform(numerical_df)
```

```python
scaled_numerical
```

```python
scaled_numerical.shape
```

```python
scaled_numerical_final = pd.DataFrame(scaled_numerical, columns = ['A1', 'A2', 'A3', 'PG1', 'PE1', 'PL1', 'PN1', 'PO1', 'D0', 'D1', 'D2',
       'D3', 'D4', 'D5', 'D6', 'QTI', 'V0', 'V1', 'V10', 'V2', 'V3', 'V4',
       'V5', 'V6', 'V7', 'V9', 'TP', 'GWP'])
```

```python
scaled_numerical_final.head()
```

```python
scaled_nonan = scaled_numerical_final.fillna(scaled_numerical_final.mean())
```

```python
scaled_nonan.isna().sum() #No nan values
```

# Setting Targets

```python
median = scaled_nonan['GWP'].median()
median
```

```python
targets = np.where(scaled_nonan['GWP']>median,1,0) #Spliting targets into 0s and 1s depending on the median 
targets
```

```python
(np.sum(targets)/targets.shape[0]).round(2) #To find whether the the split aprox 50-50
```

```python
targets = pd.DataFrame(targets, columns = ['Targets'])
targets.head(3)
```

```python
numerical_with_gwp = pd.concat([scaled_nonan,targets],axis = 1)
numerical_with_gwp.head()
```

```python
numerical_with_targets = numerical_with_gwp.drop('GWP',axis = 1)
numerical_with_targets.head()
```

```python
#Checkpoint
numerical_df4 = numerical_with_targets.copy() 
```

```python
numerical_df4.head(3)
```

```python
inputs = numerical_df4.iloc[:,:-1]
#dropping Targets column
```

# Logistic Regression using only numerical data

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(inputs,targets,train_size = 0.8, random_state = 20)
```

```python
print(x_train.shape,y_train.shape)
print(x_test.shape,y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
```

```python
reg = LogisticRegression()
```

```python
reg.fit(x_train,y_train)
```

```python
print('Score prediction:',reg.score(x_train,y_train).round(3)*100)
```

We have split our main originial data into train and test. The model has an accuracy of $84.3%$ just accounting for numerical values. Now the goal is to one-hot encode categorical data and make predictions using it.
